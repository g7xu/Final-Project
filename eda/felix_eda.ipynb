```python
each_subjects_score = each_subjects_score.dropna()
```

```python
each_subjects_score
```

```python
def truncated_mean(df):
    mean_value = df.mean(axis=1)  # First mean() gets the mean of each column, second mean() gets the mean of those means
    # truncated_value = np.trunc(mean_value)
    return mean_value
```

```python
each_subjects_avg = pd.DataFrame(truncated_mean(each_subjects_score),columns=["score"])
```

```python
each_subjects_avg = each_subjects_avg.reset_index()
```

```python
each_subjects_avg.columns
```

```python
each_subjects_avg.sort_values(by="score")
```

```python
fig, ax = plt.subplots(figsize=(20,5))


seaborn_bar = sns.barplot(each_subjects_avg, x="subject",y="score",color="green")
plt.legend(title='Arthimetic mean of every subject\'s gut microbe score', loc='upper center')
plt.savefig("each_subject's_overall_score.png");
```
